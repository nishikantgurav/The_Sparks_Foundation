## Comparing Profit and Sales

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df['Sales'],color='red',label='Sales',shade=True,bw=25)
sns.kdeplot(df['Profit'],color='Blue',label='Profit',shade=True,bw=25)
plt.xlim([-100,1000])
plt.legend()
```

# Univariate Analysis

```python
# plot the graph for Sales variable
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.title('Sales_Distribution',fontsize=20)
sns.distplot(df['Sales'],color='red')

plt.subplot(1,2,2)
sns.boxplot(df['Sales'])
plt.title('Sales_Boxplot',fontsize=20)
```
